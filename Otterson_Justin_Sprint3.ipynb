```python
# Categorize the patients based on age using a for loop and if-else statements
age_categories = []

for age in df['Age']:
    if age < 60:
        age_categories.append('Young')
    elif age < 80:
        age_categories.append('Middle-aged')
    else:
        age_categories.append('Senior')

df['Age Category'] = age_categories
df.head()
```

```python
# Remove dollar signs from the ' Charges ' column and convert to float
df[' Charges '] = df[' Charges '].apply(lambda x: float(x.replace('$', '').replace(',', '')))

# Rename the ' Charges ' column to 'Charges' for consistency
df.rename(columns={' Charges ': 'Charges'}, inplace=True)

df.head()
```

# Titanic Dataset

```python
#import data
from google.colab import files
upload = files.upload()
```

```python
# display the data
df = pd.read_csv(io.BytesIO(upload['Titanic-Dataset.csv']))
df.head
```

```python
# Descriptive statistics for the numerical columns
desc_stats = df.describe()

desc_stats
```

```python
# Demonstrate the use of a For Loop and IF ELSE
# To illustrate the use of a For Loop and IF ELSE, let's categorize the ages of passengers into various age groups: Child, Teenager, Adult, and Senior.
# We'll create a new column named AgeGroup to store these categories.
```

```python
# Categorizing age using a For Loop and IF ELSE
age_groups = []

for age in df['Age']:
    if age < 13:
        age_groups.append('Child')
    elif 13 <= age < 20:
        age_groups.append('Teenager')
    elif 20 <= age < 60:
        age_groups.append('Adult')
    elif age >= 60:
        age_groups.append('Senior')
    else:
        age_groups.append('Unknown')  # For missing age values

df['AgeGroup'] = age_groups

# Displaying the updated dataset with the new AgeGroup column
df[['Age', 'AgeGroup']].head(10)
```

Using a For Loop and IF ELSE, we've categorized passengers into different age groups:

Child: Less than 13 years
Teenager: 13 to 19 years
Adult: 20 to 59 years
Senior: 60 years and above
Unknown: For passengers with missing age values

```python
#Using loc to Subset Data:
#Let's use the loc method to filter out and display only the rows of passengers who were adults and paid a fare of more than 100 units.
```

```python
# Using loc to subset data
high_fare_adults = df.loc[(df['AgeGroup'] == 'Adult') & (df['Fare'] > 100)]

high_fare_adults.head()
```

Using the loc method, we've filtered out passengers who were adults and paid a fare of more than 100 units.

```python
#Using apply to Manipulate Data:
#We can use the apply method to create a new column that indicates whether a passenger was traveling alone or with family.
#If the sum of the columns SibSp (number of siblings/spouses) and Parch (number of parents/children) is zero, then the passenger was traveling alone;
#otherwise, they were traveling with family.
```

```python
# Using apply to create a new column indicating if the passenger was traveling alone or with family
df['TravelStatus'] = df.apply(lambda row: 'Alone' if row['SibSp'] + row['Parch'] == 0 else 'With Family', axis=1)

# Displaying the updated dataset with the new TravelStatus column
df[['SibSp', 'Parch', 'TravelStatus']].head(10)
```

Using the apply method, we've created a new column TravelStatus that indicates whether a passenger was traveling alone or with family:

Alone: The passenger was traveling without any siblings, spouses, parents, or children.
With Family: The passenger was traveling with at least one family member (sibling, spouse, parent, or child).

```python
# display the data
df = pd.read_csv(io.BytesIO(upload['heart_disease_health_indicators.csv']))
df.head
```

```python
# Compute descriptive statistics for the dataset
descriptive_stats = df.describe()

descriptive_stats
```

```python
#Demonstration using For Loop and IF ELSE
#Next, let's use a for loop combined with an if-else statement to categorize the BMI into different classes:
#'Underweight', 'Normal weight', 'Overweight', and 'Obesity'.
```

```python
# Categorize BMI using a for loop and if-else

def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal weight'
    elif 24.9 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obesity'

# Apply the function to the BMI column using a for loop
bmi_categories = [categorize_bmi(bmi) for bmi in df['BMI']]

# Add the categorized BMI to the dataframe
df['BMI_Category'] = bmi_categories

# Display the first few rows to confirm the changes
df[['BMI', 'BMI_Category']].head()
```

The BMI has been successfully categorized using a for loop and if-else statements. As we can see:

A BMI of 40 is categorized as "Obesity".
A BMI of 25 is categorized as "Overweight".
A BMI of 28 is also "Overweight".
A BMI of 27 is "Overweight".
A BMI of 24 is "Normal weight".


```python
#Using loc to Subset Data
#Next, let's use the loc function to subset the data for individuals who have had a heart disease or attack and are smokers.
```

```python
# Subset data using loc for individuals with heart disease or attack and are smokers
subset_data = df.loc[(df['HeartDiseaseorAttack'] == 1) & (df['Smoker'] == 1)]

# Display the first few rows of the subset data
subset_data.head()
```

The data has been successfully subsetted using the loc function. The displayed rows represent individuals who have had a heart disease or attack and are also smokers

```python
#Using apply to Manipulate Data
#Now, let's use the apply function to create a new column that indicates the overall health risk based on a simple criteria:

#High risk if an individual has high blood pressure, high cholesterol, and is a smoker.
#Moderate risk if an individual has two out of the three conditions mentioned above.
#Low risk otherwise.
#Let's create this new column.
```

```python
# Define a function to categorize health risk based on conditions
def categorize_health_risk(row):
    conditions_count = row['HighBP'] + row['HighChol'] + row['Smoker']
    if conditions_count == 3:
        return 'High risk'
    elif conditions_count == 2:
        return 'Moderate risk'
    else:
        return 'Low risk'

# Apply the function to the dataframe
df['Health_Risk'] = df.apply(categorize_health_risk, axis=1)

# Display the first few rows to confirm the changes
df[['HighBP', 'HighChol', 'Smoker', 'Health_Risk']].head()
```

The health risk has been successfully categorized based on the given criteria:

An individual with high blood pressure, high cholesterol, and who is a smoker is categorized as "High risk".
An individual with any two out of the three conditions mentioned is categorized as "Moderate risk".
Otherwise, the individual is categorized as "Low risk".
