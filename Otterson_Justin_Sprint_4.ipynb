```python
#importing libiraries

import pandas as pd
import io
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#puts graphic directly under code
```

```python
from google.colab import files
upload = files.upload()
```

```python
# Load the dataset
data = pd.read_csv(io.BytesIO(upload['Provisional_COVID-19_Deaths_by_Sex_and_Age_20231011.csv']))
# Display the first few rows of the dataset
data.head()
```

```python
# Generate descriptive statistics for the dataset
desc_stats = data.describe()

desc_stats
```

```python

import matplotlib.pyplot as plt


# Filter out the total for all ages and all sexes

data_filtered = data[(data['Age Group'] != 'All Ages') & (data['Sex'] != 'All Sexes')]

# Plotting deaths by Age Group
plt.figure(figsize=(15, 5))
data_filtered.groupby('Age Group')['COVID-19 Deaths'].sum().plot(kind='bar', color='skyblue')
plt.title('COVID-19 Deaths by Age Group')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plotting deaths by Sex
plt.figure(figsize=(8, 5))
data_filtered.groupby('Sex')['COVID-19 Deaths'].sum().plot(kind='bar', color='salmon')
plt.title('COVID-19 Deaths by Sex')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Plotting top 10 states with the highest COVID-19 deaths
plt.figure(figsize=(15, 5))
data_filtered.groupby('State')['COVID-19 Deaths'].sum().sort_values(ascending=False).head(10).plot(kind='bar', color='lightgreen')
plt.title('Top 10 States with Highest COVID-19 Deaths')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
