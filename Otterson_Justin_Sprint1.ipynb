Basic Statistics

```python
#Here are some basic statistics from the dataset:

#Time (Duration):
  #Average duration: 417.5 minutes (approximately 7 hours)
  #Shortest audiobook: 1 minute
  #Longest audiobook: 8,595 minutes (approximately 143 hours)
#Stars (Rating):
  #Average rating: 0.768
  #Highest rating: 5.0
  #75% of the audiobooks have a rating of 0, which suggests that many books might not have been rated yet.
#Price:
  #Average price: 559 units (the currency isn't specified)
  #Lowest price: 0 (suggesting some audiobooks might be free)
  #Highest price: 7,198 units
#Ratings (Number of Ratings):
  #Average number of ratings per audiobook: 3.723
  #Most-rated audiobook: 12,573 ratings
  #75% of the audiobooks have 0 ratings, which supports the idea that many books haven't been rated yet.
  #The statistics suggest that while there are some very popular audiobooks with high ratings and many reviews, a significant portion of the catalog hasn't been rated or reviewed yet.
```

Stretch Goal 1- Find the Top Rated Audiobooks

```python
#Next, let's identify the top-rated audiobooks. We'll consider both the star rating and the number of ratings,
#to ensure we're not just looking at books with a single high rating.
```

```python
# Filter out audiobooks with less than 10 ratings to ensure reliability
filtered_df = df[df['ratings'] > 10]

# Sort by stars and ratings
top_rated_books = filtered_df.sort_values(by=['stars', 'ratings'], ascending=False).head(10)

top_rated_books[['name', 'author', 'stars', 'ratings']]
```

Stretch Goal 2- Highest rated authors analysis

```python
#Next, let's analyze the authors to see which ones have the most audiobooks and the highest average ratings.
```

```python
# Count of audiobooks by each author
author_counts = df['author'].value_counts().head(10)

# Average rating for each author
author_avg_rating = df.groupby('author')['stars'].mean()

# Top 10 authors by average rating (considering only those with more than 10 audiobooks)
top_authors_by_rating = author_avg_rating[df['author'].value_counts() > 10].sort_values(ascending=False).head(10)

author_counts, top_authors_by_rating
```
